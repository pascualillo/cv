Pasos a seguir:
1. Sube un archivo de Excel (.xlsx) con una única columna y un encabezado con los "prompts".
2. Modifica los valores de "valor_api_openai" y "quien_soy".
3. El programa devolverá, por cada fila, el prompt y el resultado de OpenAI.

```python
# Instalar la biblioteca de OpenAI si no está instalada
!pip install openai

import time

import pandas as pd
import os
import openai
from google.colab import files


valor_api_openai = "valor_api_openai"
quien_soy = "el_limite_es_tu_imaginación"


# Definir la función para llamar a la API de OpenAI
def llamar_api_openai(valor):
    try:
        openai.api_key = valor_api_openai
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "system", "content": quien_soy},
                      {"role": "user", "content": valor}]
        )
        text = response.choices[0]['message']['content'].strip()
        return text
    except Exception as e:
        print(f"Error en la llamada a la API: {str(e)}")
        return None

# Subir el archivo XLSX
archivo_xlsx = files.upload()
nombre_archivo = list(archivo_xlsx.keys())[0]

# Leer el archivo XLSX de entrada
df = pd.read_excel(nombre_archivo)

# Obtener los valores de la columna actual (puedes ajustar el nombre de la columna según tus necesidades)
valores_actuales = df['text'].tolist()

# Recorrer los valores actuales
for i, valor_actual in enumerate(valores_actuales):
    print("Inicio.\nEsperando 3 segundos para iterar")
    time.sleep(3)  # Pausa de 3 segundos

    print('Pidiendo información a la API: \n', valor_actual)
    # Llamar a la API de OpenAI
    resultado_api = llamar_api_openai(valor_actual)

    if resultado_api is not None:
        # Actualizar el DataFrame con el resultado de la API
        df.at[i, 'Resultado API'] = resultado_api
    else:
        df.at[i, 'Resultado API'] = 'ERROR'

    # Guardar el DataFrame actualizado en un nuevo archivo XLSX
    nombre_archivo_resultado = "resultado.xlsx"
    df.to_excel(nombre_archivo_resultado, index=False)
    
    # Descargar el archivo XLSX resultante
    files.download(nombre_archivo_resultado)

    print('Fin. Datos actualizados en el Excel. \n')

print("El script ha finalizado. El archivo XLSX ha sido actualizado con los resultados de la API y se ha descargado.")
```
